```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.MultiGraph()
# hard code!
G.add_edge('data viz', 'Distribution')
G.add_edge('Distribution', 'Histrogram')
G.add_edge('Distribution', 'Boxplot')
nx.draw(G, with_labels=True)
```

```python
import pandas as pd
topics = ['mpl', 'sns', 'nx', 'np', 'pd']
df = pd.DataFrame({
    'topics': topics,
    'difficulty': ['viz', 'viz', 'graph', 'number', 'data'],
})

charac = pd.DataFrame({'topics': topics,
            'group': [1, 1, 2, 3, 4]}
     )

charac = charac.set_index('topics')
charac = charac.reindex(G.nodes())
G = nx.from_pandas_edgelist(df, 'topics', 'difficulty')
nx.draw_shell(G, with_labels=True,
       node_size=5,
         alpha=0.5, linewidths=40, 
       font_size=25, node_color=charac['group'])
```

```python
data = [['mpl', 'viz', 4], ['np', 'numTool', 5] ,
       ['sns', 'viz', 3], ['nx', 'graph', 4], 
       ['pd', 'dataTool', 5]]
cols = ['library', 'kind', 'difficulty'] # kind rating is 0->5
df = pd.DataFrame(data, columns=cols, )
# df.set_index('library', inplace=True)
df.reset_index(inplace=True); df.drop(columns='index', inplace=True)
df
```

## 2. 3.

```python
G = nx.from_pandas_edgelist(df, 'library', 'kind')
nx.draw(G, with_labels=True, font_size=20, node_size=1500, 
       alpha=0.5,) 
```

## 4.

```python
df_temp = df.reindex(G.nodes())
# df
nx.draw(G, with_labels=True, font_size=20, node_size=1500, 
       alpha=0.5, node_color=df_temp['difficulty'],
       cmap=plt.cm.Blues) 


plt.show()
```
