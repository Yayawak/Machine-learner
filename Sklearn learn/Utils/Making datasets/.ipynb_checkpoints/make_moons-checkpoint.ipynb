```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

X, z = make_moons(n_samples=50, noise=.1)
plt.scatter(X[:,0], X[:,1],cmap='autumn', c=z, s=50,
           edgecolor='k')
plt.show()
```

```python
import seaborn as sns
fig, axs = plt.subplots(figsize=(10, 15), nrows=5, ncols=2)
# i_axs = 0,1,2,3,4...
# noise_i = .1,.2,.3,.4
k = 0
for i,row_ax in enumerate(axs):
    for j, ax in enumerate(row_ax):
        noise = (k)/25
        X, z = make_moons(n_samples=50, noise=noise )
        ax.set_title('Noise : {}'.format(str(noise)), fontsize=20)
        sns.set_style('darkgrid')

        sns.scatterplot(x=X[:,0], y=X[:,1], hue=z, ax=ax,
                       palette='Set1', s=150)
        k += 1
plt.subplots_adjust(0,0,1,1.5)
# print(axs) # Matrix 2x2
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier 
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

models = [DecisionTreeClassifier(), 
         RandomForestClassifier(), 
         LogisticRegression(), 
         KNeighborsClassifier()]
model_names = ['DecisionTree', 
              'RandomForest', 
              'LogisticRegression',
              'K-nearest-neighbors']

X, z = make_moons(n_samples=80, noise=.12, random_state=1)
nmesh = 200
mx, my = np.meshgrid(
    np.linspace(X[:,0].min(), X[:,0].max(), nmesh),
    np.linspace(X[:,1].min(), X[:,1].max(), nmesh)
)
mX = np.stack([mx.ravel(), my.ravel()], 1)

plt.figure(figsize=(10,6))

for i, model in enumerate(models):
    model.fit(X, z)
    mz = model.predict(mX).reshape(nmesh, nmesh)
    
    plt.subplot(221+i, 
               xlim=(X[:,0].min(), X[:,0].max()),
               ylim=(X[:,1].min(), X[:,1].max())
               )
    plt.title(str(model_names[i]), fontsize=20, pad=10)
    plt.scatter(X[:,0], X[:,1], c=z, edgecolor='k', cmap='autumn')
    plt.contourf(mx, my, mz, cmap='autumn', alpha=0.2)
plt.subplots_adjust(0,0,1,1.2)
plt.show()
    
```
