```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# x_data = np.random.randn(100) * 3 + 3
x_data = np.arange(0, 10, 0.2)
# y depends on x 
y_data = ( np.random.randn(len(x_data)) * 2 + 1.5) * x_data
# print(x_data)
# print(y_data)

sns.scatterplot(x=x_data, y=y_data, palette='Set3', )
plt.grid()
```

```python
data_pair=pd.DataFrame(list(zip(x_data, y_data)), 
                      columns=['x', 'y'])
data_pair
lmG = sns.lmplot(data=data_pair, x='x', y='y')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
x = x_data.reshape(-1, 1)
y = y_data.reshape(-1, 1)

model = LinearRegression()
model.fit(x, y)
```

```python
model.predict([[2]])
```

```python
fig, ax = plt.subplots()
sns.regplot(data=data_pair, x='x', y='y', color='g', ax=ax)

x_pred = np.linspace(-10,10, 10).reshape(-1, 1)
y_pred = model.predict(x_pred)
# y_pred
# x_pred[0,0]
def drawFocus(x_pred, y_pred):
    l_h = plt.axhline(y=y_pred)
    l_v = plt.axvline(x=x_pred)
    plt.annotate(f'( {x_pred[0].round(2)}, {y_pred[0].round(2)} )',
                xy=(x_pred, y_pred), fontsize= 5, ha='left',
                va='bottom')
# for i in range(len(y_pred)):
#     drawFocus(x_pred[i], y_pred[i])
ax.stem(y_pred, )
# plt.show()
```

```python
a = [2, 3, 5, 20, *range(15, 18)] 
x_pred = np.array(a).reshape(-1, 1)
y_pred = model.predict(x_pred)
# x_pred
# y_pred
x = x_pred.flatten()
y = y_pred.flatten()
plt.plot(x, y, color='tab:orange')
sns.scatterplot(x=x, y=y, color='g', )
```
