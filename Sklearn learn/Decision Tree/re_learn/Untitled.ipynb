```python
def passTestResult(mid_term, final_term):
    if(mid_term>60):
        if(final_term>40):
            return 1
        else:
            return 0
    else:
        if(final_term>70):
            return 1
        else:
            return 0
```

```python
def passTestResult(mid_term, final_term):
    return np.where(mid_term>60,
                   final_term>40,
                   final_term>70)
```

```python
nmesh = 100
k = np.linspace(0, 100, nmesh)
mx, my = np.meshgrid(k, k)
mz = passTestResult(mx, my)
```

```python
plt.figure(figsize=(6,5))
plt.axes(aspect=1)

plt.xlabel("Mid term", size=20)
plt.ylabel("Final term", size=20)

plt.contourf(mx, my, mz, cmap="RdYlGn")


plt.show()
```
