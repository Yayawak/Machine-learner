```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
X = [[0, 0], [2,2]]
y = [0, 1] # from labelEncoder() or factorize()
model = tree.DecisionTreeRegressor()
model = model.fit(X, y)
print(model.predict([[1,2]]))

tree.plot_tree(model)
plt.show()
```

```python
import numpy as np

# random datasets
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
# rng.rand(5, 1) * 5
y = np.sin(X).ravel()
# y
y[::5] += 3 * (0.5 - rng.rand(16)) # make interesting outliers
regr_1 = tree.DecisionTreeRegressor(max_depth=2)
regr_2 = tree.DecisionTreeRegressor(max_depth=5)
regr_1.fit(X, y)
regr_2.fit(X, y)

X_test = np.arange(0, 5, 0.01).reshape(-1, 1)
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)

plt.figure()
plt.scatter(X, y, s=20, edgecolors='k', c='darkorange', label='data')
plt.plot(X_test, y_1, color='cornflowerblue', label='max-depth=2',
        lw=2)
plt.plot(X_test, y_2, color='yellowgreen', label='max-depth=5',
        lw=2)
plt.xlabel('data')
plt.ylabel('target')
plt.title('Decision Tree Regression')
plt.legend()

plt.show()
```
