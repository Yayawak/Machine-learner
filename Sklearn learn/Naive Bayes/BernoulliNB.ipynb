```python
from sklearn.datasets import make_classification
n_samples = 300
# dummy dataset
X, y = make_classification(n_samples=n_samples, 
                          n_features=2, 
                          n_informative=2,
                          n_redundant=0)
# X, y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
X.size, y.size
```

```python
# sns.scatterplot(x=X[:,1], y=y)
plt.plot(X, y)
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame(zip(X[:,0], X[:,1], y),  columns=['fea_1', 'fea_2', 'target'])
df.head()
```

```python
sns.pairplot(df, hue='target', corner=True)
```

```python
sns.set_style('darkgrid')
# sns.scatterplot(x=df.fea_1, y=df.fea_2,
#                hue=df.target,)
plt.scatter(df.fea_1, df.fea_2, cmap=plt.cm.RdYlBu,c=df.target)
plt.show()
```

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25)
bnb = BernoulliNB(binarize=0)
bnb.fit(X_train, y_train)
bnb.score(X_train, y_train) * 100
```
