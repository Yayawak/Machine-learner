# Preprosessing

```python
# X = df.iloc[:, :-1]
X = df.drop(columns='species',)
# X
# y = df.iloc[:, -1]
y = df['species']
# y
```

# Cross validation evaluation

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

cvs = cross_val_score(model, X, y, cv=10 )

print('cross val score : {}'. format(cvs))
print('mean (%) = {:.3f}'. format(cvs.mean() * 100))
```

# Split train & test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
k = (label, count) = np.unique(y_test, return_counts=True)
print(label)
print(count)
print(k) # this k is tuple with 2 elements
```

```python
freq = np.asarray(k) # change tuple => np.array
freq
```

```python
freq = freq.T
freq
```

```python
print(freq)
```

# Train and test accuracy of model

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
model = GaussianNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print('Accuracy Score : {:.4f}'. format(accuracy_score(y_test, y_pred) * 100))
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
import scikitplot as skplot
skplot.metrics.plot_confusion_matrix(y_test, y_pred,
                                    normalize=True,
                                   cmap='Wistia',)
# skplot.metric
plt.show()
```

```python
dy = pd.DataFrame()
dy['y_test'] = y_test
dy['y_pred'] = y_pred
dy['result'] = ( y_test == y_pred)

# dy.sort_index()
dy.sample(5)
```

```python
# see True positive
result = (y_test == y_pred)

n_tp = y_test[dy.result == True].count()
print('TP : {}'.format(n_tp))

n_tf =  y_test[dy.result == False].count()
print('TF : {}'.format(n_tf))
```

```python
# see all TF
dy[dy.result == False]
```
