```python
from sklearn.datasets import make_blobs
X,z = make_blobs(n_samples=200, centers=4, 
                cluster_std=2)
from sklearn.tree import \
DecisionTreeClassifier as dt
from sklearn.naive_bayes import GaussianNB \
as gnb
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
md=KMeans(10)
plt.scatter(X[:,0], X[:,1], c=z)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, LinearRegression, \
    Perceptron, SGDClassifier
from sklearn.svm import SVC
models = [gnb(), 
          KNeighborsClassifier(),
          dt(),
          KMeans(n_clusters=4),
         RandomForestClassifier(),
          LogisticRegression(),
          SVC(),
          LinearRegression(),
          # Perceptron(),
          SGDClassifier()
         ]
model_names = [ type(md).__name__ for md in models ]
xMinMax=[X[:,0].min(),X[:,0].max()]
yMinMax=[X[:,1].min(),X[:,1].max()]
mx, my = np.meshgrid(
    np.linspace(*xMinMax,200),
    np.linspace(*yMinMax,200))
mX=np.stack([mx.ravel(), my.ravel()],1)
# cm='gist_ncar'
# cm='gist_rainbow'
cm='hsv'
plt.figure(figsize=(15,20))
# fig, axs = plt.subplots(ncols=2, nrows=2)
for idx, md in enumerate(models):
    md.fit(X,z)
    mz = md.predict(mX).reshape(200,200)

    plt.subplot(331+idx)
    plt.title(model_names[idx], fontsize=30)
    # ax = axs[idx]
    ax = plt
    ax.title(model_names[idx], fontsize=20)
    ax.scatter(X[:,0],X[:,1],
               c=z, cmap=cm, s=170,
               edgecolor='k', lw=1.5)

    ax.contour(mx,my,mz,colors='k')
    ax.contourf(mx,my,mz,alpha=.4,
                cmap=cm)
plt.subplots_adjust(0,0.1,1,1.1)
# plt.colorbar()
plt.show()
```
